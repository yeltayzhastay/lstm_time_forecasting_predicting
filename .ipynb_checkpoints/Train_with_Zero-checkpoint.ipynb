```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt
import os

mpl.rcParams['figure.figsize'] = (8, 6)
mpl.rcParams['axes.grid'] = False
```

```python
df = df.set_index('Index')
df = pd.DataFrame(df[['Turbine_vibration_2X','Turbine_vibration_2Y']], columns = ['Turbine_vibration_2X','Turbine_vibration_2Y'])
```

```python
df['Turbine_vibration_2X+1'] = df['Turbine_vibration_2X'].shift(-1)
df['Turbine_vibration_2X+1'].fillna(20, inplace=True)
```

```python
df['Turbine_vibration_2X']
```

```python
x = df.drop('Turbine_vibration_2X+1', axis = 1)
y = df['Turbine_vibration_2X+1']

y=np.reshape(y.values, (-1,1))

scaler_x = MinMaxScaler(feature_range=(0, 1))
print(scaler_x.fit(x))
scaler_y = MinMaxScaler(feature_range=(0, 1))
print(scaler_y.fit(y))
```

# Neural Link

```python
model = Sequential()
model.add(Dense(128, input_dim=2, kernel_initializer='normal', activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
# Use a custom metricfunction

model.compile(loss='mse', optimizer='sgd', metrics=['mse'])

fit = model.fit(X_train, y_train, epochs=100,  verbose=1, validation_split=0.01)
```

# Visualizations

```python
def plot_train_history(history, title):
  mpl.rcParams['figure.figsize'] = (8, 6)
  mpl.rcParams['axes.grid'] = False
    
  loss = history.history['loss']
  val_loss = history.history['val_loss']
    
  epochs = range(len(loss))

  y = []
  for i in epochs:
        y.append(0)

  plt.figure()

  plt.plot(epochs, loss, 'b', label='Training loss')
  plt.plot(epochs, val_loss, 'r', label='Validation loss')
  plt.plot(epochs, y, 'g', label='y = 0 axis')
  plt.xlabel('Epochs')
  plt.ylabel('MSE value or Losses')
    
  plt.title(title)
  plt.legend()

  plt.show()
```

```python
plot_train_history(fit,
                   'Visualization losses')
```
