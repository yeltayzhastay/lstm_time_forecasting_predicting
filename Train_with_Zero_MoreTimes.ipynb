```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt
import os

mpl.rcParams['figure.figsize'] = (8, 6)
mpl.rcParams['axes.grid'] = False
```

```python
df1 = pd.read_excel("Revised_Data_Turbine.xlsx")
```

```python
df1 = df1.set_index('Index')
df1 = pd.DataFrame(df1[['Turbine_vibration_2X','Turbine_vibration_2Y']], columns = ['Turbine_vibration_2X','Turbine_vibration_2Y'])
df1
```

```python
print(df1.columns)
```

```python
def shaper(arr, n_time = 1):
    shape_list = []
    for i in range(0, len(arr) - n_time, 1):
        time_list = []
        for j in range(n_time):
            time_list +=arr[i + j]
        shape_list.append(time_list)
        
    columns = []
    for i in range((-n_time)+1,1,1):
        columns += ["Turbine_vibration_2X" + str(i), "Turbine_vibration_2Y" + str(i)]
    
    td = pd.DataFrame(shape_list, columns = columns)
    return td
```

```python
N_TIME = 20
df = shaper(df1.values.tolist(), N_TIME)
```

```python
df['Turbine_vibration_2X+1'] = df['Turbine_vibration_2X0'].shift(-1)
df['Turbine_vibration_2X+1'].fillna(20, inplace=True)
df['Turbine_vibration_2X0']
```

```python
x = df.drop('Turbine_vibration_2X+1', axis = 1)
y = df['Turbine_vibration_2X+1']

y=np.reshape(y.values, (-1,1))

scaler_x = MinMaxScaler(feature_range=(0, 1))
print(scaler_x.fit(x))
scaler_y = MinMaxScaler(feature_range=(0, 1))
print(scaler_y.fit(y))
```

# Variants

```python
X_trans = scaler_x.transform(x)
y_trans = scaler_y.transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_trans, y_trans, test_size=0.2)
```

# Neural Link

```python
model = Sequential()
model.add(Dense(64, input_dim=N_TIME * 2, kernel_initializer='normal', activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
# Use a custom metricfunction

model.compile(loss='mse', optimizer='adam')

fit = model.fit(X_train, y_train, epochs=50,  verbose=1, validation_split=0.5)
```

# Testing and Experiments

```python
def plot_viz(x, y, prediction, title):
    mpl.rcParams['figure.figsize'] = (8, 6)
    mpl.rcParams['axes.grid'] = False
    
    labels = ['History', 'True Future', 'Model Prediction']
    marker = ['.-', 'rx', 'go']
    
    xx = x
    x1 = []
    y1 = []
    
    for i in range(0,len(xx),2):
        x1.append(xx[i])
        y1.append(xx[i+1])

    yy = y
    pre = prediction
    epochs = range(len(x1))

    plt.figure()

    plt.plot(epochs, x1, 'b', label='X vibration')
    plt.plot(epochs, y1, 'r', label='Y vibration')
    plt.plot(len(x1), yy, 'rx', label='Real data')
    plt.plot(len(x1), pre, 'go', label='Prediction')
    plt.xlabel('time')
    plt.ylabel('value')
    
    plt.title(title)
    plt.legend()

    plt.show()

for i in range(3):
    plot = plot_viz(scaler_x.inverse_transform(X_test)[i], scaler_y.inverse_transform(y_test)[i],
                    scaler_y.inverse_transform(model.predict(X_test))[i], 'Simple Neural link model')
```

# Visualizations

```python
def plot_train_history(history, title):
  mpl.rcParams['figure.figsize'] = (8, 6)
  mpl.rcParams['axes.grid'] = False
    
  loss = history.history['loss']
  val_loss = history.history['val_loss']
    
  epochs = range(len(loss))

  y = []
  for i in epochs:
        y.append(0)

  plt.figure()

  plt.plot(epochs, loss, 'b', label='Training loss')
  plt.plot(epochs, val_loss, 'r', label='Validation loss')
  plt.plot(epochs, y, 'g', label='y = 0 axis')
  plt.xlabel('Epochs')
  plt.ylabel('MSE value or Losses')
    
  plt.title(title)
  plt.legend()

  plt.show()
```

```python
plot_train_history(fit,
                   'Visualization losses')
```
